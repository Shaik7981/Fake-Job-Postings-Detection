```python
df[['min_salary', 'max_salary']] = df['salary_range'].str.split('-',n=1, expand=True)
```

```python
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce').fillna('0')
```

```python
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce'). fillna('0')
```

```python
df[['city', 'state', 'country']] = df['location'].str.split(',', expand=True, n=2)
```

```python
df.drop(['salary_range', 'location'], axis=1, inplace=True)
```

```python
df['employment_type'].fillna("Not mentioned", inplace = True)
```

```python
cols_to_fill = ['department', 'company_profile', 'description', 'benefits', 'requirements','required_experience', 'required_education', 'industry', 'function', 'city', 'state', 'country']
df[cols_to_fill] = df[cols_to_fill].fillna('Not mentioned')
```
